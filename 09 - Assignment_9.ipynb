3.Write a function which accepts a CSV which has two columns: “ActualValues” and “PredictedValues”. You need to design a confusion matrix and give values for the following:

•Accuracy

•Misclassification Rate

•True Positive Rate

•False Positive Rate

•Specificity

•Precision

•Null Error Rate

Also plot the ROC Curve

Provide your understanding for each of them in a sentence or two.

Run this function for Results.csv

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.metrics import roc_curve, auc, RocCurveDisplay

result = confusion_matrix(rs['ActualValues'], rs['PredictedValues'])
result
```

```python
TN, FP, FN, TP = result.ravel() #TN = 3097, FP = 477, FN = 3170, TP = 670
```

```python
print(accuracy_score(rs['ActualValues'], rs['PredictedValues']))
accuracy = (TP + TN) / (TP + FP + FN + TN)
print(accuracy)
```

```python
misclassification_rate = 1 - accuracy_score(rs['ActualValues'], rs['PredictedValues'])
print(misclassification_rate)
```

```python
#true positive rate or sensitivity or recall = tp/tp+fn
print(recall_score(rs['ActualValues'], rs['PredictedValues']))
print(TP/(TP+FN))
```

```python
#false positive rate or fallout - fp/fp+tn
FPR = FP/(FP+TN)
print(FPR)
```

```python
# specificity is the recall of the negative class. You can reach it just setting the pos_label paramete = tn/tn+fp
print(recall_score(rs['ActualValues'], rs['PredictedValues'],pos_label=0))
print(TN / (TN + FP))
```

```python
# precision is TP/(TP+FP)
print(precision_score(rs['ActualValues'], rs['PredictedValues']))
print(TP/(TP+FP))
```

```python
fpr, tpr, thresholds = roc_curve(rs['ActualValues'], rs['PredictedValues'])
```

```python
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='example estimator')
display.plot()
```
