9. Count the number of articles in file 'Macbeth.txt' (Articles are 'a', 'an'& 'the') and save the file as 'Macbeth-ArticlesRemoved.txt'

```python
nltk.download('punkt')
articles = ('a', 'an', 'the', 'A', 'An', 'The')
count = 0

fl = open('./Macbeth.txt', 'r+')
mac_file = fl.read()
fl.close()

file_arr = nltk.tokenize.word_tokenize(mac_file)

for i in file_arr:
  if i in articles:
    count += 1
    file_arr.remove(i)

mac_file = " ".join(file_arr)

print(count)

fl = open('./Macbeth-ArticlesRemoved.txt', 'w+')
fl.write(mac_file)
fl.close()
```

10. Remove all the punctuations from the file. (Punctuations
are .,':;#() etc.) and save as 'Macbeth-punctuationsRemoved.txt'

```python
puncts = (".",\
          ",",\
          "'",\
          ":",\
          ";",\
          "#",\
          "(",\
          ")",\
          "[",\
          "]")

fl = open('./Macbeth-ArticlesRemoved.txt', 'r+')
mac_file = fl.read()
fl.close()

file_arr = nltk.tokenize.word_tokenize(mac_file)

for i in file_arr:
  if i in puncts:
    file_arr.remove(i)

mac_file = " ".join(file_arr)

print(mac_file)
```

12. Print the total number of words present in 'names' corpora of
nltk

```python
print(names.fileids())

male = names.raw('male.txt')
female = names.raw('female.txt')
male = male.split('\n')
female = female.split('\n')
len(male) + len(female)
```

13. For each alphabet print the frequency of names present in 'names' corpora

```python
from nltk.probability import FreqDist

fdist = FreqDist()

for word in male:
  for chr in word:
    fdist[chr.lower()] += 1

for word in female:
  for chr in word:
    fdist[chr.lower()] += 1

fdist
```

14. Plot a bar graph showing frequency of names for each alphabet

```python
import pandas as pd
import matplotlib.pyplot as plt

srs = pd.Series(dict(fdist))

plt.bar(srs[:26].index.tolist(), srs[:26].values.tolist())
```

15. Add appropriate titles, X-axis labels and Y-axis labels to the above created bar graph

```python
plt.title('Frequency of letters in names corpus')
plt.xlabel('Letters')
plt.ylabel('Count')
plt.bar(srs[:26].index.tolist(), srs[:26].values.tolist())
plt.show()
```

9.	Count the number of articles in file ‘Macbeth.txt’ (Articles are  ‘a’, ‘an’ &’the’ ) and save the file as ‘Macbeth-ArticlesRemoved.txt’

```python
file=open("Macbeth.txt",'r')
content=file.read()
file.close()

allWords=content.split()
original_count=len(allWords)
print("Total Number of Words= ",original_count)

for i,each in enumerate(allWords):
    if each.lower()=="a" or each.lower()=="an" or each.lower()=="the":
        del allWords[i]

count_after_articles_removal=len(allWords)
print("Total Number of Words after removing articles= ", count_after_articles_removal)


print("Number of Articles : " ,(original_count-count_after_articles_removal))

newfile=open('Macbeth-ArticlesRemoved.txt','w+')
newfile.write("".join(allWords))
newfile.close()
```

10.	Remove all the punctuations from the file. (Punctuations are . ,’ : ; # ( ) etc.) and save as ‘Macbeth-punctuationsRemoved.txt’

```python
file=open("Macbeth.txt",'r')
content=file.read()
file.close()

import string
list_of_punctuations=list(string.punctuation)

print("No of Characters in string : ",len(content))

for each in list_of_punctuations:
    content=content.replace(each,"")

print("No of Characters in string after removing punctuations: ",len(content))
newfile=open('Macbeth-punctuationsRemoved.txt','w+')
newfile.write(content)
newfile.close()
```

13.	For each alphabet print the frequency of names present in ‘names’ corpora

```python
import string
allAlphabets=list(string.ascii_uppercase)
```

```python
frequency=[]
for each in allAlphabets:
    wordswithEach=list(filter(lambda x:x.startswith(each),names.words()))
    frequency.append(len(wordswithEach))

for every in zip(allAlphabets,frequency):
    print(every)
```

14.	Plot a bar graph showing frequency of names for each alphabet


```python
import matplotlib.pyplot as plt
ticks=range(1,len(allAlphabets)+1)
plt.bar(ticks,frequency)
plt.xticks(ticks,allAlphabets)
plt.show()
```
